## Observação: 
. Foram tratados os dados núlos do dataset original e criado um dataset que será utilizado para análise daqui em diante (dataset_analise)

. Coluna **Propriedade**: Os dados núlos foram preenchidos com "Outros", pois verificarei nas próximas análises se são úteis ou não esses dados;

. Coluna **motivo_emp**: Os dados núlos foram preenchidos com "Outros", pois verificarei nas próximas análises se são úteis ou não esses dados;

. Coluna **pontuacao_emp**: Os dados núlos foram excluidos do dataset, tendo em vista que o proprietário dos dados não preencheu a lacuna e não temos dados suficientes para julgar a pontuação de crédido da pessoa; 

. Coluna **percentual_emp_salario**: Foi criada uma nova coluna, onde foi recalculado todos os dados com arredondamento de 2 casas, substituindo os dados núlos. O Cálculo a ser feito era a renda percentual entre salário anual x valor total solicitado. Os núlos restantes desses dados foram excluídos, tendo em vista que, ou o salário anual do cliente, ou o valor total solicitado estavam núlos e não serviriam para as análises futuras;

. Coluna **inadimplencia_anterior**: Os dados núlos foram excluidos do dataset, em vista que havia histórico de solicitações de crédito anterior dos clientes, contudo a base de dado não informava se havia ou não sido inadimplênte.

. Por fim foi criado um **Novo Dataset** onde usarei para os futuros tratamentos.
